```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/movie_ratings.csv")
```

### Q1) What's the breakdown of genres for the movies that were rated?

```python
df_genres = df['genres']
df_genres.value_counts()
genre_counts = df['genres'].value_counts().reset_index()
genre_counts.columns = ['genre', 'count']
plt.figure(figsize=(16, 6))
sns.barplot(data=genre_counts, x='genre', y='count')
plt.tight_layout()
plt.show()
```

### Q2) Which genres have the highest viewer satisfaction (highest ratings)?

```python
average_ratings = df.groupby('genres')['rating'].mean()
ratings_count = average_ratings.sort_values(ascending=False).reset_index()
ratings_count.columns = ['genre', 'average_rating']
plt.xlim(3.0, 4.0)
sns.barplot(ratings_count, y='genre', x='average_rating')
plt.tight_layout()
plt.show()
```

### Q3) How does mean rating change across movie release years?

```python
year_highest_ratings = df.groupby('year')['rating'].mean()
year_ratings_count = year_highest_ratings.sort_values(ascending=False).reset_index()
year_ratings_count.columns = ['release_year', 'average_rating']
plt.figure(figsize=(14, 6))
sns.lineplot(year_ratings_count, x='release_year', y='average_rating')
```

### Q4a) What are the 5 best-rated movies that have at least 50 ratings?

```python
average_ratings = df.groupby('title')['rating'].mean()
ratings_counts = df['title'].value_counts()
at_least_50 = ratings_counts[ratings_counts >= 50].index
selection = average_ratings.index.isin(at_least_50)
top_5_with_50_ratings = average_ratings[selection].sort_values(ascending=False)

top5 = top_5_with_50_ratings.head(5).reset_index()
top5.columns = ['title', 'average_rating']
plt.figure(figsize=(18, 4))
plt.ylim(4.2, 4.5)
sns.barplot(top5, x='title', y='average_rating')
```

### Q4b) What are the 5 best-rated movies that have at least 150 ratings?

```python
average_ratings = df.groupby('title')['rating'].mean()
ratings_counts = df['title'].value_counts()
at_least_150 = ratings_counts[ratings_counts >= 150].index
selection = average_ratings.index.isin(at_least_150)
top_5_with_150_ratings = average_ratings[selection].sort_values(ascending=False)

top5 = top_5_with_150_ratings.head(5).reset_index()
top5.columns = ['title', 'average_rating']
plt.figure(figsize=(18, 4))
plt.ylim(4.2, 4.5)
sns.barplot(top5, x='title', y='average_rating')
```
